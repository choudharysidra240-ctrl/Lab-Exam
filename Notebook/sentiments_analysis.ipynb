```python
# Cell 1: Imports and Setup
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import json
from pathlib import Path
import warnings
warnings.filterwarnings('ignore')

# Cell 2: Load Config
with open('../config.json', 'r') as f:
    config = json.load(f)
    
random_seed = config['random_seed']
np.random.seed(random_seed)

# Cell 3: Load Dataset
data_path = Path('../data/raw/legal_notices.csv')
df = pd.read_csv(data_path)

# Display first 10 rows
print("First 10 rows of dataset:")
print(df.head(10))

# Basic info
print("\nDataset Info:")
print(df.info())

# Cell 4: Class Distribution
class_counts = df['label'].value_counts()
print("\nClass Distribution:")
print(class_counts)

# Plot class distribution
plt.figure(figsize=(8, 6))
class_counts.plot(kind='bar', color=['#FF6B6B', '#4ECDC4', '#45B7D1'])
plt.title('Class Distribution of Legal Notices', fontsize=14, fontweight='bold')
plt.xlabel('Category')
plt.ylabel('Number of Samples')
plt.xticks(rotation=0)
plt.tight_layout()
plt.savefig('../results/class_distribution.png', dpi=300, bbox_inches='tight')
plt.show()

# Cell 5: Text Length Analysis
df['text_length'] = df['text'].apply(len)
print(f"\nTotal Samples: {len(df)}")
print(f"Missing Values: {df.isnull().sum().sum()}")
print(f"Average Text Length: {df['text_length'].mean():.2f} characters")
print(f"Text Length Range: {df['text_length'].min()} - {df['text_length'].max()} characters")

# Plot text length histogram
plt.figure(figsize=(10, 6))
plt.hist(df['text_length'], bins=20, edgecolor='black', alpha=0.7, color='#95D5B2')
plt.axvline(df['text_length'].mean(), color='red', linestyle='--', label=f"Mean: {df['text_length'].mean():.0f}")
plt.axvline(df['text_length'].median(), color='blue', linestyle='-.', label=f"Median: {df['text_length'].median():.0f}")
plt.title('Distribution of Legal Notice Lengths', fontsize=14, fontweight='bold')
plt.xlabel('Text Length (characters)')
plt.ylabel('Frequency')
plt.legend()
plt.tight_layout()
plt.savefig('../results/text_length_distribution.png', dpi=300, bbox_inches='tight')
plt.show()

# Cell 6: Data Quality Issues
print("\nData Quality Issues Identified:")
print("1. Potential class imbalance:", class_counts.min() / class_counts.max() < 0.8)
print("2. Text length variance:", df['text_length'].std() > 100)
print("3. Contains HTML tags or special characters:", df['text'].str.contains('<|>|&').any())
print("4. Empty or very short documents:", (df['text_length'] < 20).sum())
```
